### این فصل موارد زیر را پوشش می‌دهد:

- حل مسائل یادگیری ماشین در دنیاری واقعی در قالب یک قاب (فریم)  مورد بررسی قرار می‌گیرند
- گام‌هایی که برای توسعه یک مدل نیاز است بررسی می‌شود. 
- توزیع و نگهداری مدل بعد از توزیع آن بررسی می‌شود.

**مقدمه**

- همیشه داده برچسب‌دار در دسترس نیست و در مجموعه داده‌های ماژول کراس موجود نیست.
- بنابراین باید داده‌هایی را جمع‌آوری کنیم
- در این موارد باید مراقب بحث‌های اخلاقی و حریم خصوصی باشیم
  - نمی‌توانیم بر اساس خروجی یک الگوریتم یادگیری ماشین که ممکن است بایاس شده باشد یا تحت تاثیر کسی که برچسب می‌زند قرار گرفته باشد و یا متاثر از مجموعه داده اولیه هست، به افراد برچسب بزنیم.

 
- مراحل کلی یک مسئله یادگیری ماشین را می‌توان در سه قاب خلاصه کرد
  1. تعریف مسئله
    - در اینجا باید به تعریف ورودی، انتظار مشتری از برنامه کاربردی و معیار موفقیت روش بپردازیم.
    - باید معلوم کنیم که مشتری از برنامه‌ی ما چه چیزی می‌خواهد؟
  2. توسعه مدل
    - آماده سازی داده و ساخت مدل که می‌تواند بیش برازش کند و سپس تلاش برای تعمیم آن با استفاده از چابک سازی و در ادامه تعیین پروتکلی برای ارزیابی مدل ساخته شده در این مرحله انجام می‌شود.
  3. توزیع مدل
    - ساخت وب سرویس برنامه موبایل یا سایر چیزها برای جمع‌آوری بیشتر داده‌ها و ساخت مدل بعدی و جدید‌تر

## تعریف مسئله
- قدم اول از ذینفعال باید تا حد امکان اطلاعات کافی را کسب کنیم.
- سوالاتی را باید از ذینفعان بپرسیم؟ 
  - آیا داده‌های ورودی برچسب خورده هستند؟
  - آیا اصولاً  الگوریتم‌های یادگیری ماشین  لازم هستند؟
  - راه حل‌های موجود برای حل مسئله چطور است؟ شاید تا به حال روش‌هایی به صورت دستی بوده یا الگوریتم‌های خاصی استفاده می‌شده است.
  - چه محدودیت‌هایی داریم مثلاً آیا تاخیر شبکه در ارسال جواب جایز است یا اینکه برنامه قرار است بر روی یک سیستم جاسازی شده Embedded  اجرا شود.

### فرض‌های اصلی در حل مسئله با یادگیری عمیق
در اینجا مسئله را بر اساس فرض‌های زیر بنا می‌کنیم: 
1. داده‌های ورودی می‌توانند ما را به خروجی برسانند 
2. داده‌های مفیدی برای توصیف خروجی جمع‌آوری شده است.
  - مثلا نمی‌توانیم انتظار داشته باشیم که قیمت سهام در بورس را فقط بر اساس گذشته قیمت آن سهم پیش بینی کنیم زیرا قیمت آن به عوامل دیگری نیز کاملا وابسته است.

### جمع‌آوری داده‌ها
در مرحله تعریف مسئله باید مشخص شود که آیا مجبور به جمع‌آوری داده هستیم یا نه. این کار پرهزینه‌ترین بخش حل یک مسئله‌ی یادگیری ماشین است زیرا ممکن است تعداد بسیار زیادی ورودی نیاز داشته باشیم و همچنین به دلیل مشکلات حریم خصوصی به همه داده‌ها دسترسی نداشته باشیم.

**نکته:** ارزش داده‌ی خوب بسیار بیشتر از الگوریتم خوب است! اگر به مدت زمان تکمیل پروژه شما ۵۰ ساعت اضافه شد آن را صرف جمع‌آوری داده‌های بهتر کنید.

- آیا قرار است خودتان داده‌ها را به صورت دستی برچسب بزنید یا از سرویس‌های مانند Mechanical Turk  از آمازون که ارزان ولی نویزی هستند استفاده کنید؟
- آیا یک متخصص مانند پزشک برای برچسب زدن نیاز دارید؟
- آیا خودتان از برچسب‌ها سر در می‌آورید؟ اگر این طور باشد می‌توانید در مراحل بعدی  برای مهندسی ویژگی اقدام کنید.
- نتایج برچسب گذاری چگونه ذخیره می‌شوند؟ 

- اصل اساسی در جمع‌آوری داده این است که داده‌های جمع‌آوری شده نماینده خوبی از داده‌های اصلی باشند
  - مثلاً اگر قرار باشد از روی عکس غذا اسم غذا را تشخیص دهیم باید تمام حالت‌های ممکن را برای این عکس در نظر بگیریم (محیط کنترل‌نشده) زیرا عدم رعایت این موضوع ما را به سمت جهنم یادگیری ماشین پیش می‌برد! 
- هم چنین مشکلی به نام دریفت drift داریم یعنی داده‌های واقعی در طول زمان تغییر الگو می‌دهند و ذائقه مشتریان  به مرور تغییر می‌کند. چنین مواردی در توصیه موزیک و ارزیابی نظرات مشتریان می‌تواند موثر باشد.
- معمولاً فرض اینکه بتوانیم یک چیز را در آینده بر اساس گذشته پیش‌بینی کنیم در عمل همیشه درست نیست.

### بایاس در نمونه‌برداری
مشکل دیگر بایاس در نمونه‌برداری است. 
    در صورتی که در انتخابات ریاست جمهوری از افراد خاصی نظرسنجی شود ممکن است نتیجه واقعی با نتیجه پیش‌بینی شده خیلی متفاوت باشد.

### درک داده‌های ورودی
یکی از مهم‌ترین توصیه‌هایی که در شروع کار وجود دارد این است که داده‌ها را درک کنید.
- در صورت لزوم آنها را رسم کنید.
- آیا داده‌های شما خصیصه با مقدار ناموجود دارند؟
- آیا برچسب داده‌ها به صورت غیر بالانس است؟
- معیار موفقیت در یک مسئله چیست؟
- برای درک مطلب بهتر است به نمونه مسابقاتی که در کگل Kaggle برگزار می‌شود مراجعه کنید.

## توسعه مدل
قدم دوم توسعه مدل است.
گرچه بیشتر کتاب‌های یادگیری عمیق بر روی توسعه مدل تمرکز دارند، اما این تنها مرحله حل یک مسئله در یادگیری عمیق در دنیای واقعی نیست و نسبت به مرحله‌ی قبل ساده‌تر است.

در مورد این مرحله در فصل قبل تا حدی صحبت کردیم. دیدیم که: 
  1. باید اول داده‌ها را برداری vectorize کنیم، 
  2. سپس آنها را نرمال‌سازی کنیم یعنی به محدوده صفر تا یک انتقال دهیم یا در صورتی که چندین خصیصه داریم، آنها را نرمال کنیم.  
    - در غیر این صورت ممکن است گرادیان‌ها خیلی بزرگ شوند و موجب عدم‌همگرایی مدل شود.
  3. در مورد مقادیر ناموجود تصمیم‌گیری می‌شود:
    - مثلاً اگر خصیصه مورد نظر به صورت دسته‌ای (categorical) باشد،  مقدار نا‌موجود را به عنوان یک دسته جدید در نظر می‌گیریم. 
    - اگر داده‌های ما عددی باشند، میانگین یا میانه مقادیر آن خصیصه را به عنوان مقدار جایگزین استفاده می‌کنیم.  
  4. یکی از پروتکل‌های ارزیابی مدل را انتخاب می‌کنیم. 
     * باید دقت داشته باشیم که  مجموعه اعتبارسنجی حتی‌امکان نماینده خوبی برای داده‌های تست باشد
       -  افزونگی نداشته باشیم.
       -  اشتراک داده‌ای بین آموزش و آزمون نداشته باشیم.
  5. در مرحله پنجم باید بتوانیم به نتیجه مربوط به خط پایه غلبه کنیم برای این منظور می‌توانیم از مهندسی ویژگی، تغییر در معماری مدل، تغییر در تابع خطا، تغییر اندازه دسته Batch Size، و تغییر نرخ یادگیری استفاده کنیم.  
     * برای مثال  معیار ROC  مستقیماً نمی‌تواند به عنوان تابع خطا بهینه شود ولی می‌توانیم از یک تابع Proxy مثلا CrossEntropy  استفاده کنیم.

## توزیع مدل: 
در قدم سوم باید مدل ساخته شده را توزیع کنیم.

-  دقت داشته باشید که برای موفقیت هم باید محصول خوب تولید کنید و همین که مخاطبین شما انتظار معقولی داشته باشند. 
-  معمولا انتظار‌ها غیر‌واقعی هستند! زیرا «حس عمومی» Common Sense  مربوط به یک انسان در یک سیستم وجود ندارد.  
-  برای اینکه مشتری را متوجه این موضوع کنید توصیه می‌شود مواردی از شکست سیستم خود را به او نشان دهید!  
-  همچنین به جای اعلام یک عدد به عنوان دقت Accuracy، به صورت دقیق و مصداقی  در مورد نرخ مثبت کاذب و نرخ منفی کاذب صحبت کنید تا مشتری بتواند توانایی مدل شما را با کسب و کار خود  گره بزند.  
-  گاهی این مقادیر وابسته به مقدار آستانه‌ای هستند که باید توسط ذینفعان تنظیم شود.  
-  
  
  در ادامه این مرحله باید مدل استنتاج Inference  را به محیط مقصد منتقل کنیم که  الزاما مبتنی برPython نیست.  
  نکته مهم در این مرحله این است که قرار بر تغییر وزن‌ها نیست (آموزشی در کار نیست) و بنابراین می‌توان مدل محاسباتی شبکه را برای اجرای سریع‌‌تر، بهینه‌سازی کرد. 

#### محیط توزیع
- ممکن است بخواهیم مدل را به عنوان یک REST API  تحت وب توزیع کنیم که در این صورت باید مراقب میزان تاخیر در حد ۵۰۰ میلی‌ثانیه باشیم. 
  - مثلا در کاربردهای واقعیت مجازی VR، تاخیر ناشی از ارسال اطلاعات به سرور قابل قبول نیست. 
  - همچنین باید نگران عدم امکان ارسال داده‌های شخصی و حساس بر روی بستر وب باشیم.
- در این حالت ممکن است بخواهیم مدل استنتاج را  بر روی یک  وسیله مستقل سوار کنیم 
  - ممکن است منجر به کاهش دقت شود 
  - ولی در عوض داده‌ها به سرور ارسال  نمی‌شوند.  
  - برای مثال می‌توانیم از مدل خود بر روی تلفن همراه با استفاده از TF Lite  استفاده کنیم.  
- همچنین ممکن است مدل را بر روی مرورگر Browser  اجرا کنیم.
  - محاسبات در سمت مشتری انجام می‌شود
  - نیازی به اتصال به اینترنت نیست. 
  - نهایتا هزینه سرور کاهش پیدا می‌کند.  
  - در این حالت باید مدل سبک باشند 
  - حاوی اطلاعات محرمانه نباشد. 
  - می‌توانید از ابزارهای TF.js و WebKeras استفاده کنید.

### بهینه‌سازی مدل
برای بهینه سازی مدل استنتاج می‌توانیم از هرس وزن‌ها Weight Pruning  استفاده کنیم یا مقادیر اعشاری مربوط به وزن‌ها را را با مقادیر صحیح جایگزین کنیم  Weight Quantization  
- در این حالت اجرای مدل سبک تر می‌شود (مثلا با جایگزینی float که ۳۲ بیتی هست با int8 که ۸ بیتی هست، اندازه مدل یک چهارم می‌شود)، 
- ولی اندکی از دقت آن کم می‌شود.

- در این حالت بررسی مواردی که مدل ما به درستی عمل نکرده است، حائز اهمیت است.
